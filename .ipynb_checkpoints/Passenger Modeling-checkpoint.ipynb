# Map Starting and Ending Locations to a Bus Stop

```python
# Helper distance functions()
EPS = 50
def dist(lat1, lon1, lat2, lon2):
    '''
    returns distance in meters
    '''
    lat1, lon1, lat2, lon2 = map(float, [lat1, lon1, lat2, lon2])
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c * 1000
    return distance

def near(locA, locB):
    return abs(dist(locA[0], locA[1], locB[0], locB[1])) < EPS

def nearBusStops(lat, long):
    return [stop for stop, loc in busStopLocations.items() if near(loc, (lat, long))]

def nearestBusStop(lat, long):
    temp = [stop for stop, loc in busStopLocations.items() if near(loc, (lat, long))]
    if len(temp):
        best = temp[0]
        stoplat, stoplong = busStopLocations[best] 
        bestDist = dist(stoplat, stoplong, lat, long)
        for stop in temp[1:]:
            stoplat, stoplong = busStopLocations[stop] 
            d = dist(stoplat, stoplong, lat, long)
            if d < bestDist:
                best = stop
                bestDist = d
        return best
    else:
        return ""
```

```python
df_sessions['start'] = df_sessions.apply(lambda r: nearestBusStop(r["begin.gps_info.coords.lat"],
                                                                             r["begin.gps_info.coords.lon"]), axis=1)
df_sessions['end'] = df_sessions.apply(lambda r: nearestBusStop(r["end.gps_info.coords.lat"],
                                                                  r["end.gps_info.coords.lon"]), axis=1) 
```

```python
df_trips = df_sessions[(df_sessions['start']!='') & (df_sessions['end']!='')]
```

```python
# timebining helper functions
def createTimebin(row, timeWidth, col):
    d = parse(row[col])
    return int(d.timestamp() // timeWidth)

def timeBin(df, width):
    col = 'begin.sys_time'
    df['timebin'] = df.apply (lambda row: createTimebin(row, width, col),axis=1)
    return df

def timeBinToDate(x, width):
    return datetime.datetime.fromtimestamp(x * width)
```

```python
df_binned = timeBin(df_trips, 10 * 60)
```
